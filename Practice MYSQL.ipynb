```python
import mysql.connector as conn
import csv
mydb=conn.connect(host='localhost',user='root',passwd='root')
cursor=mydb.cursor()
cursor.execute("use cardataset")
cursor.execute("create table car(buying VARCHAR(10),maint VARCHAR(10),doors VARCHAR(10),persons VARCHAR(10),lug_boot VARCHAR(6),safety VARCHAR(5),class VARCHAR(10))")
```

```python
with open('cartable.data','r') as f:
    glass_data=csv.reader(f,delimiter='\n')
    for i in glass_data:
        p=i[0].split(',')
        print(p)
        cursor.execute(f'insert into cardataset.car values{tuple(p)}')
    mydb.commit()
```

```python
cursor.execute("select count(buying),buying from car group by buying")
cursor.fetchall()
```
